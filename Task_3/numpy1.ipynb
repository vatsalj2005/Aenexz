# NumPy & Pandas - Data Analysis with the Tips Dataset

This notebook covers NumPy arrays, vectorized operations, and data analysis using the Seaborn tips dataset.

## 1. NumPy Arrays vs Python Lists

```python
import numpy as np

# NumPy arrays enforce a single dtype
arr = np.array([1, 2, 'a', 8, 5])
print('Array:', arr)
print('Type:', type(arr))
print('dtype:', arr.dtype)  # all elements cast to string (<U11)
```

```python
# Python lists can hold mixed types
list1 = [1, 2, 'a', True]
print('Python list (mixed types):', list1)
```

## 2. Vectorized Operations

NumPy supports element-wise arithmetic — Python lists do not.

```python
# Python list: + concatenates, does not add
l1 = [1, 23, 5, 6]
print('List + [10] (concatenation):', l1 + [10])

# NumPy array: + adds element-wise
arr = np.array([1, 23, 4, 5])
print('NumPy array + 10 (element-wise):', arr + 10)
```

## 3. Boolean Masking / Filtering

```python
arr = np.array([1, 23, 4, 5])

# Boolean mask
print('Mask (arr < 5):', arr < 5)

# Filter elements where condition is True
print('Elements < 5:', arr[arr < 5])
```

## 4. Loading and Exploring the Tips Dataset

```python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
df.head()
```

```python
# Shape, info, and summary statistics
print('Shape:', df.shape)
print()
df.info()
print()
print(df.describe())
```

## 5. NumPy Statistical Functions on DataFrame Columns

```python
# Convert columns to NumPy arrays
total_bill = df['total_bill'].to_numpy()
tip = df['tip'].to_numpy()

print('Mean total_bill  :', np.mean(total_bill))
print('Median total_bill:', np.median(total_bill))
print('Std total_bill   :', np.std(total_bill))
```

## 6. Adding Derived Columns with NumPy

```python
# Calculate tip percentage (rounded to 2 decimal places)
df['tip_percentage'] = np.round((df['tip'] / df['total_bill']) * 100, 2)

# Categorise bills as High or Low relative to mean
df['bill_category'] = np.where(df['total_bill'] > np.mean(total_bill), 'High', 'Low')

df.head()
```

## 7. Filtering with Multiple Conditions

```python
# Single condition filter
print('Rows with tip_percentage > 20:')
print(df[df['tip_percentage'] > 20].shape[0], 'rows')
```

```python
# Multiple condition filter: tip > 20% AND bill is High
result = df[(df['tip_percentage'] > 20) & (df['bill_category'] == 'High')]
print('Rows with tip > 20% AND High bill:')
result
```

## 8. GroupBy Aggregation

```python
# Average tip by smoker status
print('Average tip by smoker status:')
print(df.groupby('smoker', observed=False)['tip'].mean())
```
