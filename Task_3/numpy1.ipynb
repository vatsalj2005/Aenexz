# NumPy & Pandas - Analysis on Titanic Dataset

This notebook demonstrates NumPy array operations, vectorized computation, statistical functions, filtering, and groupby — applied to the Titanic dataset.

## 1. NumPy Arrays vs Python Lists

```python
import numpy as np

# NumPy enforces a single dtype — mixing types causes upcasting
arr = np.array([1, 2, 'a', 8, 5])
print('Array:', arr)
print('dtype:', arr.dtype)  # all cast to string

# Python lists can hold mixed types
list1 = [1, 2, 'a', True]
print('Python list:', list1)
```

## 2. Vectorized Operations

```python
# Python list: + means concatenation
l1 = [1, 23, 5, 6]
print('List + [10]:', l1 + [10])   # appends

# NumPy array: + means element-wise addition
arr = np.array([1, 23, 4, 5])
print('Array + 10:', arr + 10)      # adds 10 to each element
```

## 3. Boolean Masking

```python
arr = np.array([1, 23, 4, 5])
print('Boolean mask (arr < 5):', arr < 5)
print('Elements < 5:', arr[arr < 5])
```

## 4. Load Titanic Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset.csv')

# Fill missing Age with mean so numpy operations work cleanly
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)

print('Shape:', df.shape)
df.head()
```

## 5. NumPy Statistical Functions

```python
# Convert columns to NumPy arrays
age = df['Age'].to_numpy()
fare = df['Fare'].to_numpy()

print('--- Age ---')
print('Mean  :', np.mean(age))
print('Median:', np.median(age))
print('Std   :', np.std(age))
print('Min   :', np.min(age))
print('Max   :', np.max(age))

print('\n--- Fare ---')
print('Mean  :', np.mean(fare))
print('Median:', np.median(fare))
print('Std   :', np.std(fare))
```

## 6. Adding Derived Columns with NumPy

```python
# Fare category: High or Low relative to mean
df['fare_category'] = np.where(df['Fare'] > np.mean(fare), 'High', 'Low')

# Age group: Child (<18), Adult (18-60), Senior (>60)
df['age_group'] = np.where(df['Age'] < 18, 'Child',
                   np.where(df['Age'] <= 60, 'Adult', 'Senior'))

# Round age to nearest integer
df['Age'] = np.round(df['Age']).astype(int)

df[['Name', 'Age', 'age_group', 'Fare', 'fare_category']].head(10)
```

## 7. Filtering with Single and Multiple Conditions

```python
# Single condition: female passengers
females = df[df['Sex'] == 'female']
print('Female passengers:', len(females))
```

```python
# Multiple conditions: female passengers in 1st class who survived
result = df[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]
print('1st class female survivors:', len(result))
result[['Name', 'Age', 'Fare', 'fare_category']].head()
```

```python
# High fare passengers who did NOT survive
high_fare_lost = df[(df['fare_category'] == 'High') & (df['Survived'] == 0)]
print('High-fare passengers who did not survive:', len(high_fare_lost))
```

## 8. GroupBy Aggregations

```python
# Survival rate by passenger class
print('Survival rate by Pclass:')
print(df.groupby('Pclass', observed=False)['Survived'].mean().round(2))
```

```python
# Average fare and age by Sex
print('Average fare and age by Sex:')
print(df.groupby('Sex')[['Fare', 'Age']].mean().round(2))
```

```python
# Survival count by age group
print('Survival count by age group:')
print(df.groupby('age_group')['Survived'].sum())
```
