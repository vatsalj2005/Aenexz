## 1. Install & Import Pandas

```python
import pandas as pd
print('Pandas version:', pd.__version__)
```

## 2. Pandas Series

A Series is a one-dimensional labeled array.

```python
# Series from a list with custom index
a = [1, 3, 5, 6]
print('Type of list:', type(a))
var = pd.Series(a, index=['w', 'x', 'y', 'z'])
print(var)
print('Type of Series:', type(var))
```

```python
# Series from a dictionary
calorie = {'day1': 360, 'day2': 560, 'day3': 340}
mycal = pd.Series(calorie)
print(mycal)
print('Index:', mycal.index)
```

```python
# Changing the index of a Series
mycal.index = ['mon', 'tue', 'wed']
print(mycal)
print('Updated Index:', mycal.index)
```

## 3. Pandas DataFrame

A DataFrame is a two-dimensional labeled data structure (like a table).

```python
# Creating a DataFrame from a dictionary
data = {
    'name': ['savi', 'avi', 'rohan', 'abhi'],
    'marks': [60, 70, 98, 89],
    'age': [19, 18, 20, 17]
}
df = pd.DataFrame(data)
df
```

## 4. Indexing: `[]`, `loc`, `iloc`

```python
# Select a column using []
print(df['age'])
```

```python
# loc - label-based: select row 0 with specific columns
print(df.loc[[0], ['name', 'marks']])
```

```python
# loc - select multiple rows by label
print(df.loc[[0, 1, 2]])
```

```python
# iloc - integer position-based: select row at index 2
print(df.iloc[2])
```

```python
# iloc - select multiple rows at once
print(df.iloc[[1, 2]])
```

```python
# iloc - select all rows but specific columns
print(df.iloc[:, [0, 2]])   # all rows, columns 0 and 2
print(df.iloc[0:2, [0, 1]]) # rows 0-1, columns 0 and 1
```

```python
# Rename columns
df.columns = list('abc')
print('Renamed columns:', df.columns)
df
```

## 6. Data Cleaning

### 6a. Handling Missing Values

```python
# Check for null values
print('Null counts per column:')
print(df.isnull().sum())
```

```python
# Drop rows with any null values (creates new df, does not modify original)
df_dropped = df.dropna()
print('Rows after dropna:', len(df_dropped))
```

```python
# Fill missing Calories with the mean value
df1 = pd.read_csv('record.csv')
mean_calories = df1['Calories'].mean()
df1['Calories'].fillna(mean_calories, inplace=True)
print('Null counts after filling:')
print(df1.isnull().sum())
```

### 6b. Fixing Wrong Data Format (Dates)

```python
df1 = pd.read_csv('record.csv')

# Convert Date column to datetime, coerce invalid entries to NaT
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')

# Format as DD/MM/YYYY
df1['Date'] = df1['Date'].dt.strftime('%d/%m/%Y')
print(df1[['Date']].head(10))
```

### 6c. Fixing Wrong / Outlier Data

```python
df1 = pd.read_csv('record.csv')

# Fix a specific wrong value directly
df1.loc[7, 'Duration'] = 45
print('Row 7 after fix:', df1.loc[7, 'Duration'])

# Replace all Duration values > 120 with 45 using a loop
for x in df1.index:
    if df1.loc[x, 'Duration'] > 120:
        df1.loc[x, 'Duration'] = 45

print('Max Duration after fix:', df1['Duration'].max())
```

### 6d. Removing Duplicate Rows

```python
df = pd.read_csv('record.csv')

print('Duplicates before removal:')
print(df.duplicated())

# Drop duplicate rows in place
df.drop_duplicates(inplace=True)

print('\nDuplicates after removal:')
print(df.duplicated())
```
