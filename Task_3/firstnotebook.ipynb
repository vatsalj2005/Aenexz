# Pandas with Titanic Dataset

This notebook demonstrates Pandas concepts — Series, DataFrames, indexing, and data cleaning — using the Titanic dataset from Kaggle.

## 1. Import Pandas

```python
import pandas as pd
print('Pandas version:', pd.__version__)
```

## 2. Pandas Series

A Series is a one-dimensional labeled array.

```python
# Series from a list with custom index
survival_counts = [549, 342]
s = pd.Series(survival_counts, index=['Did not survive', 'Survived'])
print(s)
print('Type:', type(s))
```

```python
# Series from a dictionary
class_counts = {'First': 216, 'Second': 184, 'Third': 491}
class_series = pd.Series(class_counts)
print(class_series)
print('Index:', class_series.index)
```

```python
# Changing the index
class_series.index = ['1st', '2nd', '3rd']
print(class_series)
```

## 3. Load Dataset into a DataFrame

```python
df = pd.read_csv('dataset.csv')
print('Shape:', df.shape)
df.head()
```

```python
# First and last rows
print('--- First 5 rows ---')
print(df.head())
print('\n--- Last 5 rows ---')
print(df.tail())
```

## 4. Indexing: `[]`, `loc`, `iloc`

```python
# Select a single column using []
print(df['Age'].head(10))
```

```python
# loc - label-based: select specific rows and columns by name
print(df.loc[[0, 1, 2], ['Name', 'Age', 'Survived']])
```

```python
# iloc - integer position-based: select by row/column number
print(df.iloc[0])           # first row
print()
print(df.iloc[[0, 1, 2]])   # multiple rows
```

```python
# iloc - all rows, specific columns
print(df.iloc[:, [0, 1, 2, 4]].head())   # PassengerId, Survived, Pclass, Sex
print()
# iloc - slice rows and specific columns
print(df.iloc[0:5, [1, 4, 5]])            # rows 0-4, Survived, Sex, Age
```

## 5. Data Cleaning

### 5a. Check for Missing Values

```python
print('Missing values per column:')
print(df.isnull().sum())
```

### 5b. Handle Missing Values

```python
# Fill missing Age with the mean age
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
print('Missing Age values after fill:', df['Age'].isnull().sum())
```

```python
# Fill missing Embarked with the most frequent value (mode)
most_common_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_common_embarked, inplace=True)
print('Missing Embarked after fill:', df['Embarked'].isnull().sum())
```

```python
# Drop the Cabin column — too many missing values (687 out of 891)
df.drop(columns=['Cabin'], inplace=True)
print('Columns after dropping Cabin:', df.columns.tolist())
```

```python
# Confirm no more nulls
print('Remaining null values:')
print(df.isnull().sum())
```

### 5c. Fix Wrong Data / Outliers

```python
# Check Fare outliers
print('Max Fare:', df['Fare'].max())
print('Rows with Fare > 300:')
print(df[df['Fare'] > 300][['Name', 'Pclass', 'Fare']])
```

```python
# Cap Fare outliers at the 99th percentile
fare_cap = df['Fare'].quantile(0.99)
for x in df.index:
    if df.loc[x, 'Fare'] > fare_cap:
        df.loc[x, 'Fare'] = fare_cap

print('Max Fare after capping:', df['Fare'].max())
```

### 5d. Fix Wrong Data Format

```python
# Pclass should be a category, not an integer
print('Pclass dtype before:', df['Pclass'].dtype)
df['Pclass'] = df['Pclass'].astype('category')
print('Pclass dtype after:', df['Pclass'].dtype)
```

### 5e. Remove Duplicate Rows

```python
print('Duplicates before removal:', df.duplicated().sum())
df.drop_duplicates(inplace=True)
print('Duplicates after removal:', df.duplicated().sum())
print('Final shape:', df.shape)
```
